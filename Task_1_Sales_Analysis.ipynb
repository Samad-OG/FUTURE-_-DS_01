```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# 1. Professional Data Loading
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year-Month'] = df['Order Date'].dt.to_period('M').astype(str)

# 2. Interactive Sales Trend (Pro Line Chart)
monthly_sales = df.groupby('Year-Month')['Sales'].sum().reset_index()
fig1 = px.line(monthly_sales, x='Year-Month', y='Sales',
              title='üöÄ Business Revenue Growth (Monthly Trend)',
              template='plotly_dark', markers=True)
fig1.update_traces(line_color='#00d4ff')
fig1.show()

# 3. Top 10 Products with Profit Analysis (Pro Bar Chart)
top_p = df.groupby('Product Name').agg({'Sales':'sum', 'Profit':'sum'}).sort_values(ascending=False, by='Sales').head(10).reset_index()
fig2 = px.bar(top_p, x='Sales', y='Product Name', color='Profit',
             title='üèÜ Top 10 Products: Sales vs Profitability',
             orientation='h', template='plotly_dark',
             color_continuous_scale='Viridis')
fig2.show()

# 4. Regional Performance Map (Donut Chart)
region_data = df.groupby('Region')['Sales'].sum().reset_index()
fig3 = px.pie(region_data, values='Sales', names='Region', hole=0.5,
             title='üåç Market Share by Region',
             color_discrete_sequence=px.colors.sequential.RdBu,
             template='plotly_dark')
fig3.update_traces(textposition='inside', textinfo='percent+label')
fig3.show()
```
